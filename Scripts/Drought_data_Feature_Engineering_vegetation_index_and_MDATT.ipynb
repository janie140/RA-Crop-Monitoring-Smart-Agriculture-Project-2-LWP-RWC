```python
import pandas as pd
import numpy as np
import torch as t
```

```python
# load the dataset
drought_df = pd.read_excel('drought_training_df.xlsx')

drought_df
```

```python
# Create a new df called indices_df by copying df
indices_df = drought_df.copy()
indices_df = drought_df.iloc[:, 1:] #keep bands only
indices_df
```

```python
# Round Certain Wavelengths to the Nearest Whole Number to create Indices

wavelengths = indices_df.columns[0:].astype(int)
def find_nearest(value):
  return str(wavelengths[np.abs(wavelengths-value).argmin()])
```

# Create Existing Vegetation Indices (Non Optimized)

```python

# Mutate a new column, Disease Water Stress Index (DWSI) (x802 -x547)/(x1657+x682)
indices_df["DWSI"] = (indices_df[find_nearest(802)] -indices_df[find_nearest(547)])/(indices_df[find_nearest(1657)] + indices_df[find_nearest(682)])

# Mutate a new column, Moisture Stress Index (MSI) x1599/x819
indices_df["MSI"] = indices_df[find_nearest(1599)]/indices_df[find_nearest(819)]

# Mutate a new column, Leaf Water Vegetation Index 1 (LWVI1) (x1094 -x983)/(x1094+x983)
indices_df["LWVI1"] = (indices_df[find_nearest(1094)] -indices_df[find_nearest(983)])/(indices_df[find_nearest(1094)] + indices_df[find_nearest(983)])

# Mutate a new column, Leaf Water Vegetation Index 2 (LWVI2) (x1094 -x1205)/(x1094+x1205)
indices_df["LWVI2"] = (indices_df[find_nearest(1094)] -indices_df[find_nearest(1205)])/(indices_df[find_nearest(1094)] + indices_df[find_nearest(1205)])

# Mutate a new column, Normalized Difference Infrared Index (NDII) (x819 -x1649)/(x819+x1649)
indices_df["NDII"] = (indices_df[find_nearest(819)] -indices_df[find_nearest(1649)])/(indices_df[find_nearest(819)] + indices_df[find_nearest(1649)])

# Mutate a new column, Normalized Difference Water Index (NDWI) (x857 -x1241)/(x857+x1241)
indices_df["NDWI"] = (indices_df[find_nearest(857)] -indices_df[find_nearest(1241)])/(indices_df[find_nearest(857)] + indices_df[find_nearest(1241)])

# Mutate a new column, Water Band Index (WBI) x970/x900
indices_df["WBI"] = indices_df[find_nearest(970)]/indices_df[find_nearest(900)]

# Mutate a new column, Normalized Multi-band Drought Index (NMDI)(x860 -(x1640 -x2130))/(x860+(x1640 -x2130))
indices_df["NMDI"] = (indices_df[find_nearest(860)] - (indices_df[find_nearest(1640)] -indices_df[find_nearest(2130)])) / (indices_df[find_nearest(860)] + (indices_df[find_nearest(1640)] -indices_df[find_nearest(2130)]))

indices_df
```

```python
target = drought_df.iloc[:, 0]
target
```

```python
bands_df = drought_df.copy()
# Drop the first column
bands_df = bands_df.iloc[:, 1:]
bands_df
```

# **Optimized** Vegetation Indices for this dataset


```python
###### Determine the columns for SD Index

# Create the find_optimal_indices function
def find_optimal_indices(data, target):
    num_columns = data.shape[1]
    best_correlation = 0
    best_indices = None

    for i in range(num_columns - 1):
        for j in range(i + 1, num_columns):
            subtracted_data = data.iloc[:, i] - data.iloc[:, j]
            correlation = np.corrcoef(subtracted_data, target)[0, 1]
            if abs(correlation) > best_correlation:
                best_correlation = abs(correlation)
                best_indices = (i, j)

    return best_indices

# Find the best indices
best_indices_SD = find_optimal_indices(bands_df, target)

# Convert best_indices tuple to a list
best_indices_list = list(best_indices_SD)

# Get the column names corresponding to the best indices
column_names = bands_df.columns[best_indices_list]

# Print the results
print("Best indices:", best_indices_SD)
print("Corresponding column names:", column_names)

# Mutate a new column, SD which is the columns at best_indices_list[0] - best_indices_list[1]
indices_df["SD"] = indices_df.iloc[:, best_indices_list[0]] - indices_df.iloc[:, best_indices_list[1]]
```

```python
# Determine the columns for SR Index

# Create the find_optimal_indices function
def find_optimal_indices(data, target):
    num_columns = data.shape[1]
    best_correlation = 0
    best_indices = None

    for i in range(num_columns - 1):
        for j in range(i + 1, num_columns):
            divided_data = data.iloc[:, i] / data.iloc[:, j]
            correlation = np.corrcoef(divided_data, target)[0, 1]
            if abs(correlation) > best_correlation:
                best_correlation = abs(correlation)
                best_indices = (i, j)

    return best_indices

# Find the best indices
best_indices_SR = find_optimal_indices(bands_df, target)

# Convert best_indices tuple to a list
best_indices_list = list(best_indices_SR)

# Get the column names corresponding to the best indices
column_names = bands_df.columns[best_indices_list]

# Print the results
print("Best indices:", best_indices_SR)
print("Corresponding column names:", column_names)

# Mutate a new column, SR which is the columns at best_indices_list[0] / best_indices_list[1]
indices_df["SR"] = indices_df.iloc[:, best_indices_list[0]] / indices_df.iloc[:, best_indices_list[1]]
```

```python
# Determine the columns for ND Index

# Create the find_optimal_indices function
def find_optimal_indices(data, target):
    num_columns = data.shape[1]
    best_correlation = 0
    best_indices = None

    for i in range(num_columns - 1):
        for j in range(i + 1, num_columns):
            calculated_data = np.abs(data.iloc[:, i] - data.iloc[:, j]) / (data.iloc[:, i] + data.iloc[:, j])
            correlation = np.corrcoef(calculated_data, target)[0, 1]
            if abs(correlation) > best_correlation:
                best_correlation = abs(correlation)
                best_indices = (i, j)

    return best_indices

# Find the best indices
best_indices_ND = find_optimal_indices(bands_df, target)

# Convert best_indices tuple to a list
best_indices_list = list(best_indices_ND)

# Get the column names corresponding to the best indices
column_names = bands_df.columns[best_indices_list]

# Print the results
print("Best indices:", best_indices_ND)
print("Corresponding column names:", column_names)

# Mutate a new column, ND which is the columns at |best_indices_list[0] âˆ’ best_indices_list[1]|/(best_indices_list[0] + best_indices_list[1])
indices_df["ND"] = np.abs(indices_df.iloc[:, best_indices_list[0]] - indices_df.iloc[:, best_indices_list[1]]) / (indices_df.iloc[:, best_indices_list[0]] + indices_df.iloc[:, best_indices_list[1]])
```

```python
# Please select Runtime > Change Runtime Type > T4 GPU
DEVICE = 'cuda:0' if t.cuda.is_available() else 'cpu:0'

#Parse hypespectral bands (bands) and Water Potential (Mpa) (target) from dataset
df_np = drought_df.to_numpy()
target = drought_df['Water Potential (Mpa)'].to_numpy()
new_target = t.from_numpy(target.astype(np.float32)).to(device=DEVICE)
bands = t.from_numpy(df_np[:, 1:].astype(np.float32)).to(device=DEVICE)

#Customized correlation coefficient function
def corrcoef(X, y):
  """
  X is (n, m) array
  with m samples and n possible mdatt indices

  y is (1, m) array with m samples and
  """
  mu_X = X.mean(axis=-1, keepdims=True)
  mu_y = y.mean(axis=-1, keepdims=True)
  Xcent = X - mu_X # X centered around origin
  ycent = y - mu_y # y centered around origin
  n = X.shape[-1]
  C_ij = (Xcent * ycent).sum(axis=-1) / (n - 1)
  C_ii = (Xcent * Xcent).sum(axis=-1) / (n - 1)
  C_jj = (ycent * ycent).sum(axis=-1) / (n - 1)
  return C_ij / ((C_ii * C_jj).sqrt())

#Function that finds the best MDATT indices based on the absolute value of corrcoef
def _find_mdatt_indices(bands, new_target):
  best_correlation = -np.inf

  #Vectorization of bands for increased efficiency
  bands_i = bands[:, :, np.newaxis, np.newaxis]
  bands_j = bands[:, np.newaxis, :, np.newaxis]
  bands_k = bands[:, np.newaxis, np.newaxis, :]

  B = 50 # Batch size
  MAXB = 993 # Max bands
  for i in range(0, MAXB, B):
    end_i = min(i+B, MAXB)
    for j in range(0, MAXB, B):
      end_j = min(j+B, MAXB)
      for k in range(0, MAXB, B):
        end_k = min(k+B, MAXB)
        this_band_i = bands_i[:, i:end_i, :, :]
        this_band_j = bands_j[:, :, j:end_j, :]
        this_band_k = bands_k[:, :, :, k:end_k]
        bands_kj_diff = this_band_k - this_band_j
        bands_kj_diff = t.where(
            bands_kj_diff == 0,
            1e-4, bands_kj_diff)
        mdatt = (this_band_k - this_band_i) / bands_kj_diff
        mdatt_shape = mdatt.shape # (53,65,65,65)
        correlation = corrcoef(mdatt.reshape(53, -1).T, new_target).abs() # (65^3, 53) , (1, 53)
        
        correlation[correlation.isnan()] = 0 # Remove nans
        i_j_k_ravelled = (correlation).cpu().numpy().argmax() # convert to numpy
        bi, bj, bk = np.unravel_index(i_j_k_ravelled, mdatt_shape[1:]) #getting the best indices within the batch
        x = correlation[i_j_k_ravelled]
        if x > best_correlation:
          best_correlation = x
          best_indices = (bi+i, bj+j, bk+k) #saving the absolute index value of the best mdatt bands
            
  return best_indices

def find_mdatt_indices(*args, **kw):
  with t.no_grad(): # disable autogradients
    return _find_mdatt_indices(*args, **kw)
```

```python
# Find the best indices
best_indices_MDATT = _find_mdatt_indices(bands, new_target)

# Convert best_indices tuple to a list
best_indices_list = list(best_indices_MDATT)

# Get the column names corresponding to the best indices
column_names = bands_df.columns[best_indices_list]

# Print the results
print("Best indices:", best_indices_MDATT)
print("Corresponding column names:", column_names)

# Mutate a new column, MDATT which is (RÎ»3 âˆ’ RÎ»1)/(RÎ»3 âˆ’ RÎ»2)
indices_df["MDATT"] = (indices_df.iloc[:, best_indices_list[2]] - indices_df.iloc[:, best_indices_list[0]]) / (indices_df.iloc[:, best_indices_list[2]] - indices_df.iloc[:, best_indices_list[1]])
```
