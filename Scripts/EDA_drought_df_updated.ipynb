```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# load the dataset
drought_df = pd.read_excel('drought_training_df.xlsx')

drought_df
```

```python
drought_df.hist("Water Potential (Mpa)") #heavily left-skewed
```

# Correlation between spectral bands and Water Potential

```python
# Create a correlation matrix for all columns of spectral bands and WP
corr_matrix = drought_df.corr()
corr_matrix
```

```python
# Only keep the correlations with the target variable
wp_corrs = corr_matrix["Water Potential (Mpa)"]
wp_corrs
```

```python
# Remove the 1st entry in wp_corrs
wp_corrs = wp_corrs.drop(wp_corrs.index[0])
wp_corrs
```

```python
import matplotlib.pyplot as plt

# Increase the figure size
fig, ax = plt.subplots(figsize=(10, 10))

# Plot wp_corrs as a horizontal bar chart
wp_corrs.plot(kind='barh', ax=ax)

# Add labels and title
ax.set_xlabel('Correlation')
ax.set_ylabel('Hyperspectral Band from 339nm to 2515nm in 994 bins (lowest at bottom)')
ax.set_title('Water Potential Correlation vs Bands')

# Rotate y-axis labels by 45 degrees
ax.set_yticklabels(wp_corrs.index, rotation=45)

# Remove x-axis tick marks
ax.set_yticks([])

# Adjust spacing between subplots and labels
plt.tight_layout()

# Show the plot
plt.show()
```

```python
# Get the top 20 absolute correlation indices
top_20_indices = wp_corrs.abs().sort_values(ascending=False).head(20).index

# Select columns from drought df based on the top indices and keep the first column
top_corr_d = drought_df[drought_df.columns[:1].tolist() + top_20_indices.tolist()]

top_corr_d
```

```python
# save file for top 20 absolute correlation indices
top_corr_d.to_excel('top_20_corr_d.xlsx', index = False
                    
```

```python
# plot one specified hyperspectral band value vs Water Potential

def plot_band_vs_wp(i, order=2, data = drought_df):
    
    # Construct the variable name
    band_variable = str(i)
    
    # Check if the variable exists in the DataFrame
    if band_variable not in data.columns:
        raise ValueError(f"Variable '{band_variable}' does not exist in the DataFrame.")
    
    # Plot the specified hyperspectral band vs Water Potential with best fit line
    sns.regplot(x=band_variable, y='Water Potential (Mpa)', order=order, data=data)
    
    # Set axis labels
    plt.xlabel('Hyperspectral band value')
    plt.ylabel('Water Potential (Mpa)')
    
    # Set plot title
    plt.title(f'Water Potential (Mpa) vs Hyperspectral Band {i}')
    
    # Show the plot
    plt.show()

# Example usage
plot_band_vs_wp(2501)


plot_band_vs_wp(339)
```

```python
# Showing 16 Images of bands from top 20 corr bands

# Define the bands you want to plot
bands_to_plot = ['1894', '1898', '1891', '1902', '1887', '1988', '1993', '1990', '1985', '1982', '1979', '1996',
                 '2001', '1998', '2004', '1977']

# Create a figure and subplots
fig, axes = plt.subplots(nrows=4, ncols=4, figsize=(16, 8), subplot_kw={})

# Flatten the axes array for easier iteration
axes_flat = axes.flatten()

# Plot each band on a separate subplot
for band, ax in zip(bands_to_plot, axes_flat):
    sns.regplot(x=band, y='Water Potential (Mpa)', order=2, data=drought_df, ax=ax)
    ax.set_xlabel(f'{band} value')

plt.tight_layout()
plt.show()
```
