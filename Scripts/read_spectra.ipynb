```python
!pip install specdal
from specdal import Collection, Spectrum, read
import pandas as pd
import numpy as np
import os
```

```python
# load the hyperspectral data folder
datadir = "/content/drive/MyDrive/Drought_Experiment/Old Data/Spectral Data (Handheld SVC Spectroradiometer)/"
df = pd.DataFrame()

# read all files within a block and get a dataframe consists of all readings
for file_date in os.listdir(datadir):
  datadir1 = os.path.join(datadir, file_date)

  try: # handle non .sig file
    for block in os.listdir(datadir1):
      datadir2 = os.path.join(datadir1, block)
      spectrum = Collection(name='test_spectrum')
      for f in os.listdir(datadir2):
        s = Spectrum(filepath=os.path.join(datadir2, f))
        spectrum.append(s)

      spectrum_tranposed = spectrum.data.T # transpose the dataframe
      spectrum_tranposed = spectrum_tranposed.reset_index(names='file_name')
      # spectrum_tranposed['Treatment'] = np.where(spectrum_tranposed['file_name'].str.split('[_-]').str[0] =='B', 'biochar', 'no biochar')
      spectrum_tranposed['Drought Status'] = np.where(spectrum_tranposed['file_name'].str.split('[_-]').str[1] =='D', 'drought', 'no drought')
      spectrum_tranposed['Genotype'] = spectrum_tranposed['file_name'].str.slice(-1)
      spectrum_tranposed['Block'] = block.split(' ')[-1].replace("Block","") # since there are folders named Block3 instead of Block 3, we have to take care of corner cases
      spectrum_tranposed['Date'] = file_date.replace('_','/')
      spectrum_tranposed.drop(['file_name'], axis=1, inplace=True)
      df = pd.concat([df, spectrum_tranposed]) # concatenante all files into a big dataframe
  except:
    continue

df = df[['Date', 'Drought Status', 'Genotype', 'Block'] + [c for c in df if c not in ['Date', 'Drought Status', 'Genotype', 'Block']]] # reposition columns
df.reset_index(drop=True, inplace=True)
df
```

```python
# load the excel file of WP
file_path = '/content/drive/MyDrive/Drought_Experiment/Old Data/Water Potential Data (Clean).xlsx'
wp_data = pd.read_excel(file_path)
wp_data
```

```python
# convert data types in 2 dfs preparing to join
df['Drought Status'] = df['Drought Status'].astype(str)
# df['Treatment'] = df['Treatment'].astype(str)
df['Genotype'] = df['Genotype'].astype(str)
df['Block'] = df['Block'].astype(str)
df['Date'] = pd.to_datetime(df['Date'])

wp_data['Drought Status'] = wp_data['Drought Status'].astype(str)
# wp_data['Treatment'] = wp_data['Treatment'].astype(str)
wp_data['Genotype'] = wp_data['Genotype'].astype(str)
wp_data['Block'] = wp_data['Block'].astype(str)
wp_data['Date'] = pd.to_datetime(wp_data['Date'])
```

```python
# merge the spectral and water potential dataframes on 4 columns 'Date', 'Drought Status', 'Genotype', 'Block'
join_columns = ['Date', 'Drought Status', 'Genotype', 'Block']
wp_merged_df = pd.merge(df, wp_data, on=join_columns)

print(wp_merged_df)
```

```python
# check missing data
wp_merged_df.isnull().sum()
```

```python
# move Water Potential column to the column 5
column_to_move = 'Water Potential (Mpa)'
target_position = 4

columns = wp_merged_df.columns.tolist()
columns.remove(column_to_move)
columns.insert(target_position, column_to_move)

wp_merged_df = wp_merged_df[columns]

print(wp_merged_df)
```

```python
# Round the column names of the wavelength bands to integers and convert to strings
wp_merged_df.columns = [str(round(col)) if isinstance(col, float) else col for col in wp_merged_df.columns]
wp_merged_df
```

```python
# save the original file
wp_merged_df.to_excel('wp_merged_df_original.xlsx', index = False)
```
