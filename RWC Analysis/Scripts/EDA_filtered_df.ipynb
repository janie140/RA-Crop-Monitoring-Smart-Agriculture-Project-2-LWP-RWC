```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# load the dataset
drought_df = pd.read_excel('filtered_training_df.xlsx')

drought_df
```

```python
drought_df.hist("RWC") #heavily left-skewed
```

# Correlation between spectral bands and RWC

```python
# Create a correlation matrix for all columns of spectral bands and RWC
corr_matrix = drought_df.corr()
corr_matrix
```

```python
# Only keep the correlations with the target variable 'RWC'
rwc_corrs = corr_matrix["RWC"]
rwc_corrs
```

```python
# Remove the last 2 entries in rwc_corrs
rwc_corrs = rwc_corrs[:-2]

rwc_corrs
```

```python
import matplotlib.pyplot as plt

# Increase the figure size
fig, ax = plt.subplots(figsize=(10, 10))

# Plot wp_corrs as a horizontal bar chart
rwc_corrs.plot(kind='barh', ax=ax)

# Add labels and title
ax.set_xlabel('Correlation')
ax.set_ylabel('Hyperspectral Band from 339nm to 2515nm in 994 bins (lowest at bottom)')
ax.set_title('Relative Water Content Correlation vs Bands')

# Rotate y-axis labels by 45 degrees
ax.set_yticklabels(rwc_corrs.index, rotation=45)

# Remove x-axis tick marks
ax.set_yticks([])

# Adjust spacing between subplots and labels
plt.tight_layout()

# Show the plot
plt.show()
```

```python
# plot one specified hyperspectral band value vs RWC

def plot_band_vs_rwc(i, order=2, data = drought_df):
    
    # Construct the variable name
    band_variable = str(i)
    
    # Check if the variable exists in the DataFrame
    if band_variable not in data.columns:
        raise ValueError(f"Variable '{band_variable}' does not exist in the DataFrame.")
    
    # Plot the specified hyperspectral band vs RWC with best fit line
    sns.regplot(x=band_variable, y='RWC', order=order, data=data)
    
    # Set axis labels
    plt.xlabel('Hyperspectral band value')
    plt.ylabel('Relative Water Content')
    
    # Set plot title
    plt.title(f'Relative Water Content vs. Hyperspectral Band {i}')
    
    # Show the plot
    plt.show()

# Example usage
plot_band_vs_rwc(2501)


plot_band_vs_rwc(339)
```

```python
# Showing 16 Images of random bands 

# Define the bands you want to plot
bands_to_plot = ['494', '576', '638', '790', '836', '992', '1004', '1175', '1238', '1340', '1568', '1673',
                 '1745', '1884', '1977', '2507']

# Create a figure and subplots
fig, axes = plt.subplots(nrows=4, ncols=4, figsize=(16, 8), subplot_kw={})

# Flatten the axes array for easier iteration
axes_flat = axes.flatten()

# Plot each band on a separate subplot
for band, ax in zip(bands_to_plot, axes_flat):
    sns.regplot(x=band, y='RWC', order=2, data=drought_df, ax=ax)
    ax.set_xlabel(f'{band} value')

plt.tight_layout()
plt.show()
```
