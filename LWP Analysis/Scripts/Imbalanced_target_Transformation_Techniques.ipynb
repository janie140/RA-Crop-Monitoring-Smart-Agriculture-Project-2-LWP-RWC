```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# load the dataset
drought_df = pd.read_excel('drought_training_df_without_outliers.xlsx')

drought_df
```

```python
drought_df.hist("Water Potential (Mpa)") #heavily left-skewed
```

```python
# Cube Root Transformation
drought_df["Water Potential (Mpa)_cbrt"] = np.cbrt(drought_df["Water Potential (Mpa)"])

drought_df.hist("Water Potential (Mpa)_cbrt") 
```

```python
drought_df_target_cbrt_transform = drought_df.drop(columns = 'Water Potential (Mpa)')

drought_df_target_cbrt_transform.head()
```

```python
# save new training file
drought_df_target_cbrt_transform.to_excel('drought_training_cbrt_transform.xlsx', index = False)
```

```python
# Yeo-Johnson Transformation
# Similar to Box-Cox but more flexible as it can handle both positive and negative values
from scipy.stats import yeojohnson
transformed_data, lambda_value = yeojohnson(drought_df["Water Potential (Mpa)"])

# Plot histogram of transformed data
plt.hist(transformed_data, bins=20, color='blue', alpha=0.7)
plt.title('Histogram of Transformed Water Potential')
plt.xlabel('Transformed Water Potential')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
drought_df_target_Yeo_Johnson_transform = drought_df.drop(columns = 'Water Potential (Mpa)')
drought_df_target_Yeo_Johnson_transform['Water Potential (Mpa)_YJ_transform'] = transformed_data

drought_df_target_Yeo_Johnson_transform.head()
```

```python
# save new training file
drought_df_target_Yeo_Johnson_transform.to_excel('drought_training_Yeo_Johnson_transform.xlsx', index = False)
```
